## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("diabetes.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,:-1].values
print(input)
output = data.iloc[:,-1].values
print(output)
```

## ***SPLLITING THE DATASET INTO TRAIN AND TEST DATA***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input,output)
```

## ***TRAINING THE DATASET***

```python
from keras.models import Sequential
from keras.layers import Dense

# designing the model
ANN_Model = Sequential()
ANN_Model.add(Dense(12, input_dim=8, activation="relu"))
ANN_Model.add(Dense(8,activation="relu"))
ANN_Model.add(Dense(1,activation="sigmoid"))

# compiling the model
ANN_Model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# training the model
ANN_Model.fit(x_train,y_train,epochs=10,batch_size=5)
```

## ***CALCULATING THE ACCURACY OF THE TRAINED MODEL***

```python
ANN_Model.evaluate(x_test,y_test)
```
